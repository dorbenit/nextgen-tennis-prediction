```python
import pandas as pd
import re

# Load data
players_path = 'atp_players.csv'
matches_path = 'challenger_1998_2025.csv'

df_players = pd.read_csv(players_path)
df_matches = pd.read_csv(matches_path)

# Convert tourney_date to datetime
df_matches['tourney_date'] = pd.to_datetime(df_matches['tourney_date'], format='%Y%m%d', errors='coerce')

# Drop rows with missing critical stat values
columns_to_check = [
    'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms',
    'w_bpSaved', 'w_bpFaced', 'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon',
    'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced', 'minutes'
]
df_cleaned = df_matches.dropna(subset=columns_to_check).copy()

# Define list of players to check
player_ids_to_check = [
    100644, 104918, 106234, 210097, 111815, 207989, 134770, 106421, 103970,
    133430, 106043, 106233, 105208, 200000, 126207, 104792, 105777, 128034,
    111202, 207733, 206173, 105668, 104542, 104545, 105223, 103781, 111575,
    105453, 104731, 106378, 207518, 106298, 106065, 105227, 126610, 104022,
    105683, 106401, 104925, 105807, 104745, 104755, 105138, 104417, 103819,
    105023, 104527, 126774, 126203, 104607, 126205
]

# Stats to average
stats_columns = ['ace', 'df', 'svpt', '1stIn', '1stWon', '2ndWon', 'SvGms', 'bpSaved', 'bpFaced']

# Function to calculate total games from score string
def total_games_in_score(score_str):
    if not isinstance(score_str, str) or any(x in score_str for x in ['RET', 'W/O', 'DEF']):
        return None
    sets = re.findall(r'(\d+)-(\d+)', score_str)
    return sum(int(a) + int(b) for a, b in sets)

# Extract player features
player_features = {}

for pid in player_ids_to_check:
    matches = df_cleaned[
        ((df_cleaned['winner_id'] == pid) & (df_cleaned['winner_age'] <= 24)) |
        ((df_cleaned['loser_id'] == pid) & (df_cleaned['loser_age'] <= 24))
    ].copy().sort_values(by='tourney_date')

    total_matches = matches.shape[0]
    if total_matches < 10:
        continue

    if matches.empty:
        continue

    first_match = matches.iloc[0]
    entry_age = first_match['winner_age'] if first_match['winner_id'] == pid else first_match['loser_age']

    wins = matches[matches['winner_id'] == pid]
    losses = matches[matches['loser_id'] == pid]

    features = {'Entry_Age': entry_age}

    # Average stats for wins
    if not wins.empty:
        for stat in stats_columns:
            features[f'Avg_Win_{stat}'] = wins[f'w_{stat}'].mean()
        features['Avg_Win_Minutes'] = wins['minutes'].mean()
        features['Avg_Win_Total_Games'] = wins['score'].apply(total_games_in_score).dropna().mean()
        # Serve %s
        features['Win_1stServePct'] = (wins['w_1stIn'] / wins['w_svpt']).mean()
        features['Win_1stWonPct'] = (wins['w_1stWon'] / wins['w_1stIn']).mean()
        features['Win_2ndWonPct'] = (wins['w_2ndWon'] / (wins['w_svpt'] - wins['w_1stIn'])).mean()
        features['Win_bpSavePct'] = (wins['w_bpSaved'] / wins['w_bpFaced']).mean()
        features['Avg_Opponent_Rank_Wins'] = wins['loser_rank'].mean()

    # Average stats for losses
    if not losses.empty:
        for stat in stats_columns:
            features[f'Avg_Loss_{stat}'] = losses[f'l_{stat}'].mean()
        features['Avg_Loss_Minutes'] = losses['minutes'].mean()
        features['Avg_Loss_Total_Games'] = losses['score'].apply(total_games_in_score).dropna().mean()
        # Serve %s
        features['Loss_1stServePct'] = (losses['l_1stIn'] / losses['l_svpt']).mean()
        features['Loss_1stWonPct'] = (losses['l_1stWon'] / losses['l_1stIn']).mean()
        features['Loss_2ndWonPct'] = (losses['l_2ndWon'] / (losses['l_svpt'] - losses['l_1stIn'])).mean()
        features['Loss_bpSavePct'] = (losses['l_bpSaved'] / losses['l_bpFaced']).mean()
        features['Avg_Opponent_Rank_Losses'] = losses['winner_rank'].mean()

    # Avg aces by surface
    surfaces = ['Hard', 'Clay', 'Grass']
    for surface in surfaces:
        surface_matches = matches[matches['surface'] == surface]
        surface_aces = []
        surface_aces += surface_matches[surface_matches['winner_id'] == pid]['w_ace'].tolist()
        surface_aces += surface_matches[surface_matches['loser_id'] == pid]['l_ace'].tolist()
        features[f'Avg_Aces_{surface}'] = pd.Series(surface_aces).mean()

    # Add height and hand
    player_row = df_players[df_players['player_id'] == pid]
    if not player_row.empty:
        features['Height'] = player_row['height'].values[0] if pd.notna(player_row['height'].values[0]) else None
        features['Hand'] = player_row['hand'].values[0] if pd.notna(player_row['hand'].values[0]) else None
    else:
        features['Height'] = None
        features['Hand'] = None

    features['reach_sf'] = 1  # LABEL
    player_features[pid] = features

# Create dataframe
df_player_features = pd.DataFrame.from_dict(player_features, orient='index')
df_player_features.index.name = 'player_id'

# Add full name
df_players['full_name'] = df_players['name_first'] + ' ' + df_players['name_last']
df_player_features = df_player_features.merge(
    df_players[['player_id', 'full_name']], left_index=True, right_on='player_id', how='left'
).rename(columns={'full_name': 'name'})

# Reorder
base_cols = ['player_id', 'name', 'Entry_Age', 'Height', 'Hand', 'reach_sf']
other_cols = [col for col in df_player_features.columns if col not in base_cols]
df_player_features = df_player_features[base_cols + other_cols]
df_player_features
```

```python
import pandas as pd
import re
import random

# Load data
players_path = 'atp_players.csv'
matches_path = 'challenger_1998_2025.csv'

df_players = pd.read_csv(players_path)
df_matches = pd.read_csv(matches_path)

# Convert date column
df_matches['tourney_date'] = pd.to_datetime(df_matches['tourney_date'], format='%Y%m%d', errors='coerce')

# Drop rows with missing stats
columns_to_check = [
    'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms',
    'w_bpSaved', 'w_bpFaced', 'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon',
    'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced', 'minutes'
]
df_cleaned = df_matches.dropna(subset=columns_to_check).copy()

# Function to calculate games from score
def total_games_in_score(score_str):
    if not isinstance(score_str, str) or any(x in score_str for x in ['RET', 'W/O', 'DEF']):
        return None
    sets = re.findall(r'(\d+)-(\d+)', score_str)
    return sum(int(a) + int(b) for a, b in sets)

# IDs to exclude
player_ids_to_check = [
    100644, 104918, 106234, 210097, 111815, 207989, 134770, 106421, 103970,
    133430, 106043, 106233, 105208, 200000, 126207, 104792, 105777, 128034,
    111202, 207733, 206173, 105668, 104542, 104545, 105223, 103781, 111575,
    105453, 104731, 106378, 207518, 106298, 106065, 105227, 126610, 104022,
    105683, 106401, 104925, 105807, 104745, 104755, 105138, 104417, 103819,
    105023, 104527, 126774, 126203, 104607, 126205
]

# Filter matches with age <= 24
matches_under_25 = df_cleaned[(df_cleaned['winner_age'] <= 24) | (df_cleaned['loser_age'] <= 24)].copy()

# Count matches per player under 25
from collections import defaultdict
match_count_dict = defaultdict(int)
for _, row in matches_under_25.iterrows():
    if row['winner_age'] <= 24:
        match_count_dict[row['winner_id']] += 1
    if row['loser_age'] <= 24:
        match_count_dict[row['loser_id']] += 1

# Filter eligible players
eligible_players = [pid for pid, count in match_count_dict.items() if count >= 50 and pid not in player_ids_to_check]

# Random sample of 100 players
random.seed(42)
sample_ids = random.sample(eligible_players, 100)

# Define stats
stats_columns = ['ace', 'df', 'svpt', '1stIn', '1stWon', '2ndWon', 'SvGms', 'bpSaved', 'bpFaced']

# Extract features
sample_features = {}

for pid in sample_ids:
    matches = matches_under_25[
        ((matches_under_25['winner_id'] == pid) & (matches_under_25['winner_age'] <= 24)) |
        ((matches_under_25['loser_id'] == pid) & (matches_under_25['loser_age'] <= 24))
    ].copy().sort_values(by='tourney_date')

    if matches.shape[0] < 50:
        continue

    first_match = matches.iloc[0]
    entry_age = first_match['winner_age'] if first_match['winner_id'] == pid else first_match['loser_age']

    wins = matches[matches['winner_id'] == pid]
    losses = matches[matches['loser_id'] == pid]

    features = {'Entry_Age': entry_age}

    # Aggregate wins
    if not wins.empty:
        for stat in stats_columns:
            features[f'Avg_Win_{stat}'] = wins[f'w_{stat}'].mean()
        features['Avg_Win_Minutes'] = wins['minutes'].mean()
        features['Avg_Win_Total_Games'] = wins['score'].apply(total_games_in_score).dropna().mean()

        # Win %s
        features['Win_1stServePct'] = (wins['w_1stIn'] / wins['w_svpt']).mean()
        features['Win_1stWonPct'] = (wins['w_1stWon'] / wins['w_1stIn']).mean()
        features['Win_2ndWonPct'] = (wins['w_2ndWon'] / (wins['w_svpt'] - wins['w_1stIn'])).mean()
        features['Win_bpSavePct'] = (wins['w_bpSaved'] / wins['w_bpFaced']).mean()
        features['Avg_Opponent_Rank_Wins'] = wins['loser_rank'].mean()

    # Aggregate losses
    if not losses.empty:
        for stat in stats_columns:
            features[f'Avg_Loss_{stat}'] = losses[f'l_{stat}'].mean()
        features['Avg_Loss_Minutes'] = losses['minutes'].mean()
        features['Avg_Loss_Total_Games'] = losses['score'].apply(total_games_in_score).dropna().mean()

        features['Loss_1stServePct'] = (losses['l_1stIn'] / losses['l_svpt']).mean()
        features['Loss_1stWonPct'] = (losses['l_1stWon'] / losses['l_1stIn']).mean()
        features['Loss_2ndWonPct'] = (losses['l_2ndWon'] / (losses['l_svpt'] - losses['l_1stIn'])).mean()
        features['Loss_bpSavePct'] = (losses['l_bpSaved'] / losses['l_bpFaced']).mean()
        features['Avg_Opponent_Rank_Losses'] = losses['winner_rank'].mean()

    # Aces by surface
    for surface in ['Hard', 'Clay', 'Grass']:
        surface_matches = matches[(matches['surface'] == surface)]
        if not surface_matches.empty:
            aces = []
            for _, row in surface_matches.iterrows():
                if row['winner_id'] == pid:
                    aces.append(row['w_ace'])
                elif row['loser_id'] == pid:
                    aces.append(row['l_ace'])
            features[f'Avg_Aces_{surface}'] = sum(aces) / len(aces) if aces else None
        else:
            features[f'Avg_Aces_{surface}'] = None

    # Add height and hand
    row = df_players[df_players['player_id'] == pid]
    if not row.empty:
        features['Height'] = row['height'].values[0] if pd.notna(row['height'].values[0]) else None
        features['Hand'] = row['hand'].values[0] if pd.notna(row['hand'].values[0]) else None
    else:
        features['Height'] = None
        features['Hand'] = None

    features['reach_sf'] = 0  # Label
    sample_features[pid] = features

# Build DataFrame
df_negative = pd.DataFrame.from_dict(sample_features, orient='index')
df_negative.index.name = 'player_id'
df_players['full_name'] = df_players['name_first'] + ' ' + df_players['name_last']
df_negative = df_negative.merge(
    df_players[['player_id', 'full_name']], left_index=True, right_on='player_id', how='left'
).rename(columns={'full_name': 'name'})

# Reorder
base_cols = ['player_id', 'name', 'Entry_Age', 'Height', 'Hand', 'reach_sf']
rest = [col for col in df_negative.columns if col not in base_cols]
df_negative = df_negative[base_cols + rest]
df_negative
```
